## Adicionar uma coluna

### A partir de uma que já existe

```python
vendas_df['Comissao'] = vendas_df['Valor Final'] * 0.05
display(vendas_df)
```

### Criar uma nova, com valor padrão

```python
vendas_df.loc[:, 'Imposto'] = 0
display(vendas_df)
```

## Excluir linhas e colunas

### Colunas

```python
vendas_df = vendas_df.drop('Imposto', axis=1)
display(vendas_df)
```

### Linhas

```python
vendas_df = vendas_df.drop(0, axis=0)
display(vendas_df)
```

## Valores Vazios

### Deletar linhas/colunas vazias

```python
vendas_df = vendas_df.dropna(how='all', axis=1)
display(vendas_df)

vendas_df = vendas_df.dropna(how='any', axis=0)
display(vendas_df)
```

### Linhas que possuem pelo menos um valor vazio

```python
vendas_df = vendas_df.dropna()
display(vendas_df)
```

## Preecher valores vazios (media e ultimo valor)

```python
vendas_df['Comissao'] = vendas_df['Comissao'].fillna(vendas_df['Comissao'].mean())
display(vendas_df)

vendas_df = vendas_df.ffill()
display(vendas_df)
```

## Calcular indicadores

### Value Counts

```python
transacoes_por_loja = vendas_df['ID Loja'].value_counts()
display(transacoes_por_loja)
```

### Groupby

#### Todas as colunas

```python
faturamento_produto = vendas_df.groupby('Produto').sum(numeric_only=True)
display(faturamento_produto)
```

#### Colunas especificas

```python
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)

faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').mean()
display(faturamento_produto)

faturamento_produto = vendas_df[['ID Loja','Produto', 'Valor Final']].groupby('ID Loja').sum(numeric_only=True)
display(faturamento_produto)
```

## Mesclar 2 dataframes

```python
gerentes_df = pd.read_excel('databases/Gerentes.xlsx')
display(gerentes_df)

vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```
